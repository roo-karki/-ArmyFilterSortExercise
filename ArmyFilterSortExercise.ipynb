```python
grped_data = army.groupby(['regiment'])['deaths','deserters','readiness','armored', 'battles',	'size',	'veterans'].sum()
grped_data
```

1. Nighthawks has significantly higher deaths compared to other two regiment.
2. Nighthawks had highest number of readiness in battles and  has most armored army.
3. Nighthawks had fought most battles.

```python
per_death = grped_data.filter(['deaths','battles'], axis=1)
per_death['death_per_battle'] = per_death['deaths']/per_death['battles']
per_death
```

4. Scouts had significantly less deaths per battle compared to other two regiment. This could be because their army size which was highest among three regiment.
